```python
# 
def gammaCorrection(image, gamma):
    new_image = image.copy()
    (w, h, d) = new_image.shape
    print(w,h)
    for i in range(w):
        for j in range(h):
            new_image[i,j] = [0,0,0]
            #new_image[i,j] = 255 * np.power(new_image[i,j] /255, 1 / gamma)
    return new_image
```
